```python
def get_image_files(directory):
    # List to hold the file data
    file_data = []

    # Walk through the directory
    for root, dirs, files in os.walk(directory):
        for file in files:
            # Check if the file is an image by its extension
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp', '.tiff', '.tif')):
                # Append the folder name and file name to the list
                folder_name = os.path.basename(root)
                file_data.append((folder_name, file))

    # Create a DataFrame from the list
    df = pd.DataFrame(file_data, columns=['Folder', 'FileName'])
    
    # Remove everything after underscore in FileName column
    df['FileName'] = df['FileName'].str.split('_').str[0]

    return df
```

```python
directory = img_path  # Replace with your directory path
df = get_image_files(directory)
```
